> 
> Подумайте, что полезно именно вам, и реализуйте — это будет лучшим решением для вашего развития.

```python
# Можно начинать творить тут
'''
Функция find_primes определяет является ли число простым или нет. Простое число делится нацело только
на 1 и себя. Функция принимает на вход список int, на выходе должна вернуть список простых чисел.
'''
def find_primes(input_nums):
    counter = 0
    output = []
    for i in input_nums:
# здесь мы проводим итерацию от 1 до самого числа        
        for j in range(1,i+1): 
            if i % j == 0:
# для каждого значения j проверяем делится ли число i нацело на j, если да - мы увеличиваем счётчик на 1 
                counter += 1 
        if counter == 2:
# если после полного прохода цикла по j, счётчик равен 2, это значит, что у числа только 2 делителя, мы добавляем это число в output 
            output.append(i)
# сбрасываем счётчик для проверки следующего числа    
        counter = 0 
    return output    
find_primes([1,2,3,4,5,11,13,17])   
```
